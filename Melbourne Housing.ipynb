```python
#Fill BuildingArea missing values using median by Rooms
buildingarea_medians = df.groupby("Rooms")["BuildingArea"].median()

def fill_building_area(row):
    if pd.isnull(row["BuildingArea"]):
        return buildingarea_medians.get(row["Rooms"], df["BuildingArea"].median())
    else:
        return row["BuildingArea"]

df["BuildingArea"] = df.apply(fill_building_area, axis=1)
```

```python
# Fill any remaining BuildingArea (e.g. rare room counts) with overall median
df["BuildingArea"] = df["BuildingArea"].fillna(df["BuildingArea"].median())
```

```python
# fill missing Car values with 0 (no car spot assumed)
df["Car"] = df["Car"].fillna(0)
```

```python
print(df.isnull().sum())
```

```python
# Fill missing CouncilArea with 'Unknown'
df["CouncilArea"] = df["CouncilArea"].fillna("Unknown")
```

```python
# Convert Date to datetime format
df["Date"] = pd.to_datetime(df["Date"], format="%d/%m/%Y", errors="coerce")
# format the datetime as a string in mm/dd/yyyy
df["Date"] = df["Date"].dt.strftime("%m/%d/%Y")
```

```python
# Convert Postcode to integer (no decimals needed for postcodes)
df["Postcode"] = df["Postcode"].astype(int)
```

```python
# Convert Propertycount to integer (counts should be whole numbers)
df["Propertycount"] = df["Propertycount"].astype(int)
```

```python
# Convert Car to integer (car spots should be whole numbers)
df["Car"] = df["Car"].astype(int)
```

```python
# Check number of duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
# Drop duplicates
df = df.drop_duplicates()
```

```python
# Add is_outlier flag
price_thresh = df["Price"].quantile(0.99)
landsize_thresh = df["Landsize"].quantile(0.99)
buildingarea_thresh = df["BuildingArea"].quantile(0.99)

df["is_outlier"] = (
    (df["Price"] > price_thresh) |
    (df["Landsize"] > landsize_thresh) |
    (df["BuildingArea"] > buildingarea_thresh)
).astype(int)

print(f"Number of outliers flagged: {df['is_outlier'].sum()}")
```
